```python
Dataset1=Dataset1[['Course Name','Difficulty Level','Course Description','Skills']]
Dataset1.head()
```

```python
Dataset1['Course Name'] = Dataset1['Course Name'].str.replace(' ',',')
Dataset1['Course Name'] = Dataset1['Course Name'].str.replace(',,',',')
Dataset1['Course Name'] = Dataset1['Course Name'].str.replace(':','')
Dataset1['Course Description'] = Dataset1['Course Description'].str.replace(' ',',')
Dataset1['Course Description'] = Dataset1['Course Description'].str.replace(',,',',')
Dataset1['Course Description'] = Dataset1['Course Description'].str.replace('_','')
Dataset1['Course Description'] = Dataset1['Course Description'].str.replace(':','')
Dataset1['Course Description'] = Dataset1['Course Description'].str.replace('(','')
Dataset1['Course Description'] = Dataset1['Course Description'].str.replace(')','')

#removing paranthesis from skills columns
Dataset1['Skills'] = Dataset1['Skills'].str.replace('(','')
Dataset1['Skills'] = Dataset1['Skills'].str.replace(')','')
```

```python
Dataset1
```

```python
Dataset1['Tags']=Dataset1['Course Name']+ Dataset1['Difficulty Level']+ Dataset1['Course Description']+Dataset1['Skills']
```

```python
Dataset1
```

```python
Dataset1['Tags'].iloc[1]
```

```python
new_df = Dataset1[['Course Name','Tags']]
```

```python
new_df.head(5)
```

```python
new_df['Tags'] = Dataset1['Tags'].str.replace(',',' ')
```

```python
new_df['Course Name'] = Dataset1['Course Name'].str.replace(',',' ')
```

```python
new_df.rename(columns = {'Course Name':'course_name'}, inplace = True)
```

```python
new_df['Tags'] = new_df['Tags'].apply(lambda x:x.lower())
```

```python
new_df.head(5)
```

```python
new_df.shape
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['Tags']).toarray()
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y=[]

    for i in text.split():
        y.append(ps.stem(i))

    return " ".join(y)
```

```python
new_df['Tags'] = new_df['Tags'].apply(stem)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
def recommend(course):
    course_index = new_df[new_df['course_name'] == course].index[0]
    distances = similarity[course_index]
    course_list = sorted(list(enumerate(distances)),reverse=True, key=lambda x:x[1])[1:7]

    for i in course_list:
        print(new_df.iloc[i[0]].course_name)
```

```python
recommend('Finance for Managers')
```

```python
recommend('Business Strategy Business Model Canvas Analysis with Miro')
```

```python
recommend('Python Programming Essentials')
```

```python
recommend('Business Statistics and Analysis Capstone')
```
